# Criando Variável de Nível Sócio Econômico

## Selecionando e Justificando as Variáveis

1. Região do Curso
2. Cor ou Raça
3. Quantidade de Pessoas Morando Juntos
4. Renda Total da Familia
5. Situação de Trabalho
6. Necessidade de Financiamento
8. Necessidade de Cotas
9. Tipo de Escola no Ensino Médio
10. Familiar com Ensino Superior

*A proxy que será proposta irá se basear principalmente na distribuição de renda para cada uma das variáveis para determinar o nível socio economicos dos alunos.*
  
*Cada variável será verificada pela média de renda e assim atribuidos pontos de 1 a 5. Ao final os alunos serão classificados em seus devidos níveis sócio econômicos.*

*Essa classificação será dada pela somatória dos pontos. Sendo a definição das divisões das classes pelo método:*

#### Regra de Sturges

# $k = 1 + \frac {10}{3}\log_{10}n$

# Código Nível Sócio Econômico

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('microdados_enade_2018econ.csv', sep=';')
```

```python
df.drop_duplicates(inplace=True)
```

## Função de nível socio economico

```python
def pontos_nivel_socio_economico_regiao(aluno):
  if aluno['Co Regiao Curso'] == 3:
    pontos = 1
  elif aluno['Co Regiao Curso'] == 4:
    pontos = 2
  elif aluno['Co Regiao Curso'] == 5:
    pontos = 3
  elif aluno['Co Regiao Curso'] == 1:
    pontos = 4
  elif aluno['Co Regiao Curso'] == 2:
    pontos = 5
  else:
    pontos = 0
  return pontos * 1

def pontos_nivel_socio_economico_cor(aluno):
  if aluno['Qe I02'] == 'A':
    pontos = 1
  elif aluno['Qe I02'] == 'D':
    pontos = 2
  elif aluno['Qe I02'] == 'C':
    pontos = 3
  elif aluno['Qe I02'] == 'E':
    pontos = 4
  elif aluno['Qe I02'] == 'B':
    pontos = 5
  else:
    pontos = 0
  return pontos * 3

def pontos_nivel_socio_economico_qtde_pessoas(aluno):
  if aluno['Qe I07'] == 'A':
    pontos = 1
  elif aluno['Qe I07'] == 'B':
    pontos = 2
  elif aluno['Qe I07'] == 'C':
    pontos = 3
  elif aluno['Qe I07'] == 'D':
    pontos = 4
  elif aluno['Qe I07'] == 'E':
    pontos = 5
  elif aluno['Qe I07'] == 'F':
    pontos = 5
  elif aluno['Qe I07'] == 'G':
    pontos = 5
  elif aluno['Qe I07'] == 'H':
    pontos = 5    
  else:
    pontos = 0
  return pontos * 1

def pontos_nivel_socio_economico_renda(aluno):
  if aluno['Qe I08'] == 'A':
    pontos = 8
  elif aluno['Qe I08'] == 'B':
    pontos = 4
  elif aluno['Qe I08'] == 'C':
    pontos = 2
  elif aluno['Qe I08'] == 'D':
    pontos = 1
  elif aluno['Qe I08'] == 'E':
    pontos = 0
  elif aluno['Qe I08'] == 'F':
    pontos = 0
  elif aluno['Qe I08'] == 'G':
    pontos = 0
  else:
    pontos = 0
  return pontos * 4

def pontos_nivel_socio_economico_trabalho(aluno):
  if aluno['Pontos Renda'] > 1:
      if aluno['Qe I10'] == 'A':
        pontos = 5
      elif aluno['Qe I10'] == 'B':
        pontos = 4
      elif aluno['Qe I10'] == 'C':
        pontos = 3
      elif aluno['Qe I10'] == 'D':
        pontos = 2
      elif aluno['Qe I10'] == 'E':
        pontos = 1
      else:
        pontos = 0
  else:
    pontos = 0
  return pontos * 2

def pontos_nivel_socio_economico_financiamento(aluno):
  if aluno['Qe I11'] == 'K':
    pontos = 5
  elif aluno['Qe I11'] == 'G':
    pontos = 4
  elif aluno['Qe I11'] == 'A':
    pontos = 2
  elif aluno['Qe I11'] == 'B':
    pontos = 1
  else:
    pontos = 3

  return pontos * 3

def pontos_nivel_socio_economico_cotas(aluno):
  if aluno['Qe I15'] == 'B':
    pontos = 1
  else:
    pontos = 5

  return pontos * 2

def pontos_nivel_socio_economico_escola(aluno):
  if aluno['Qe I17'] == 'A':
    pontos = 5
  elif aluno['Qe I17'] == 'D':
    pontos = 4
  elif aluno['Qe I17'] == 'E':
    pontos = 3
  elif aluno['Qe I17'] == 'B':
    pontos = 2
  elif aluno['Qe I17'] == 'C':
    pontos = 1
  elif aluno['Qe I17'] == 'F':
    pontos = 1
  else:
    pontos = 0

  return pontos * 2

def pontos_nivel_socio_economico_superior(aluno):
  if aluno['Qe I21'] == 'A':
    pontos = 1
  elif aluno['Qe I21'] == 'B':
    pontos = 5
  else:
    pontos = 0

  return pontos * 2
```

```python
df['Pontos Regiao'] = df.apply(pontos_nivel_socio_economico_regiao, axis=1)
df['Pontos Cor'] = df.apply(pontos_nivel_socio_economico_cor, axis=1)
df['Pontos Qtde'] = df.apply(pontos_nivel_socio_economico_qtde_pessoas, axis=1)
df['Pontos Renda'] = df.apply(pontos_nivel_socio_economico_renda, axis=1)
df['Pontos Trabalho'] = df.apply(pontos_nivel_socio_economico_trabalho, axis=1)
df['Pontos Financiamento'] = df.apply(pontos_nivel_socio_economico_financiamento, axis=1)
df['Pontos Cotas'] = df.apply(pontos_nivel_socio_economico_cotas, axis=1)
df['Pontos Escola'] = df.apply(pontos_nivel_socio_economico_escola, axis=1)
df['Pontos Superior'] = df.apply(pontos_nivel_socio_economico_superior, axis=1)
```

```python
def pontos_nivel_socio_economico(aluno):
  if aluno['Pontos Totais'] >= 80:
    pontos = 'Z'
  elif aluno['Pontos Totais'] < 80 and aluno['Pontos Totais'] >= 70:
    pontos = 'F'
  elif aluno['Pontos Totais'] < 70 and aluno['Pontos Totais'] >= 60:
    pontos = 'E'
  elif aluno['Pontos Totais'] < 60 and aluno['Pontos Totais'] >= 50:
    pontos = 'D'
  elif aluno['Pontos Totais'] < 50 and aluno['Pontos Totais'] >= 40:
    pontos = 'C'
  elif aluno['Pontos Totais'] < 40 and aluno['Pontos Totais'] >= 30:
    pontos = 'B'
  elif aluno['Pontos Totais'] < 30:
    pontos = 'A'
  else:
    pontos = None
  return pontos
```

```python
df['Nivel Socio Econ'] = df.apply(pontos_nivel_socio_economico, axis=1)
```
